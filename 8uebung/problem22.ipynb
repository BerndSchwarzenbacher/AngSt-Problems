```r
require(ggplot2)
```

```r
power <- function(lam) {
    # Return power of test in problem 22 for given lambda
    return (1 - ppois(7, 8*lam))
}
```

```r
lstep <- 0.01
lmax <- 2
lambda <- seq(0.5+lstep, lmax, by=lstep)
```

```r
df <- data.frame(lambda=lambda, power=power(lambda))
```

```r
ggplot(df, aes(lambda, power)) + geom_point()
```
